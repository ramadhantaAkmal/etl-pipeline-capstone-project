```python
def to_long_format(df:pd.DataFrame) -> pd.DataFrame:
    value_name = f'{df['Indicator Name'][0]}({df['Indicator Code'][0]})'
    dt = df.drop(columns=['Indicator Name', 'Indicator Code'])
    dt['Country Name'] = dt['Country Name'].str.strip() 
    dt.columns = ['Country_Name', 'Country_Code'] + [str(col) for col in dt.columns[2:]]
    dt_long = pd.melt(dt, id_vars=['Country_Name', 'Country_Code'], 
                   var_name='Year', value_name=value_name)
    dt_long['Year'] = dt_long['Year'].astype(int)
    return dt_long
    
```

```python
gdp_data_transformed = to_long_format(gdp_data)
gdp_data_transformed
```

```python
population_data_transformed = to_long_format(population_data)
population_data_transformed
```

```python
merged_df_a = pd.merge(gdp_data_transformed, population_data_transformed, on=['Country_Code','Country_Name', 'Year'], how='inner')
merged_df_a
```

```python
electricity_access_pct_transformed = to_long_format(electricity_access_percent)
rural_population_pct_transformed = to_long_format(rural_population_percent)
merge_df_b = pd.merge(electricity_access_pct_transformed, rural_population_pct_transformed, on=['Country_Code','Country_Name', 'Year'], how='inner')

df_final = pd.merge(merged_df_a, merge_df_b, on=['Country_Code','Country_Name', 'Year'], how='inner')
```

```python
df_final.to_csv('./world_development.csv')
```
